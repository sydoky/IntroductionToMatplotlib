```python
heart_disease = pd.read_csv('/content/heart-disease.csv')
```

```python
heart_disease.plot.hist(figsize=(10, 30), subplots=True); #10 is width and 30 is height 
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
```

```python
over_50
```

```python
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             figsize=(14,9))
```

```python
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
ax.set_xlim([45,100]);
```

```python
fig, ax = plt.subplots(figsize=(10, 6)) #object oriented method #this line we create a graph itself
scatter = ax.scatter(over_50["age"], #here we create 
                     over_50["chol"],
                     c=over_50["target"]) #c will give a color in the graph. C is a building key word c = color
ax.set(title="Heart disease and cholesterol level",
       xlabel="Age",
       ylabel="cholesterol");
ax.legend(*scatter.legend_elements(),title="Target") #*scatter is a synthex, the elements of legend should be scatter, legend_elements is a building function, target our dots

ax.axhline(over_50['chol'].mean(),
           linestyle='--')
```

```python
fig,(ax0, ax1)=plt.subplots(nrows=2,
                            ncols=1,
                            sharex=True,
                            figsize=(10,8))
scatter = ax0.scatter(over_50["age"],
                      over_50["chol"],
                      c=over_50["target"])

ax0.set(title="Heart Disease and Cholesterol Level", #ax0.set is setting a title
        ylabel="Cholestrol");
ax0.legend(*scatter.legend_elements(),title="Target");

ax0.axhline(over_50["chol"].mean(),
            color='b',
            linestyle="--",
            label="Average")

scatter=ax1.scatter(over_50["age"],
                    over_50["thalach"],
                    c=over_50['target'])

ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(),title="Target");

ax1.axhline(over_50["thalach"].mean(),
            color="r",
            linestyle="--",
            label="Average")

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")



```
